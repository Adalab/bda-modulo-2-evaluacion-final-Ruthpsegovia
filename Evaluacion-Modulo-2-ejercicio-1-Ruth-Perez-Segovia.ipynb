```python
import requests
import pandas as pd
import mysql.connector
import numpy as np
from mysql.connector import Error
```

# Fase 1:Extracción de datos de películas

```python
url = "https://beta.adalab.es/resources/apis/pelis/pelis.json"

datos = requests.get(url)
datos.status_code
datos.content

datos_pelis = datos.json()
```

# Fase 2: Creación de la Base de Datos

```python
try:
    mycursor= conexion.cursor()
    query="CREATE DATABASE IF NOT EXISTS movies_db"
    mycursor.execute(query)
    print("Query exitosa")
except:
    print("Error.")
mycursor.execute("USE movies_db")
```

```python
mycursor.execute("USE movies_db")

query_films='''CREATE TABLE IF NOT EXISTS films(
    id INT AUTO_INCREMENT PRIMARY KEY,
    title VARCHAR(150),
    year INT,
    duration INT,
    genre VARCHAR(100),
    adult_content BOOLEAN
);'''

mycursor.execute(query_films)
conexion.commit()
print("Tablas creadas exitosamente")
```

```python
df_movies_db = pd.DataFrame(datos_pelis)
df_movies_db
df_clean = df_movies_db.replace({np.nan: None, 'nan': None, 'NaN': None})
```

# Fase 3: Inserción de los Datos en la Base de Datos (QUERYS)

```python
TRY:
    cnx = connection()
    IF cnx:
        mycursor = cnx.cursor()
        mycursor.execute("USE movies_db")
        
    insert_films_query = """
        INSERT INTO films (title, year, duration, genre, adult_content)
        VALUES (%s, %s, %s, %s, %s);
        """   
        
TRY:
    cnx = connection()
    IF cnx:
        mycursor = cnx.cursor()
        mycursor.execute("USE movies_db")

        insert_films_query = """
        INSERT INTO films (title, year, duration, genre, adult_content)
        VALUES (%s, %s, %s, %s, %s);
        """

        # Recorremos cada fila del DataFrame limpio
       FOR index, row in df_clean.iterrows():
            movie_data = (
                row["title"],    
                row["year"],      
                row["duration"],  
                row["genre"],    
                row["adult_content"]    
            )

            mycursor.execute(insert_films_query, movie_data)
            print(f"Insertada: {row['titulo']}")

        cnx.commit()
        print(f"{len(df_clean)} películas insertadas exitosamente")

        mycursor.close()
        cnx.close()

EXCEPT Error AS e:
    print("Error:", e)
```
